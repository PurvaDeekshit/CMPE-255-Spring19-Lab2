```python
#Function to calculate mean of numbers passed in the list
def mean_num_friends(x):
    # TODO
    print("List: ", x)
    sum_of_all_elements = sum(x)
    print("Sum of all elements in list: ", sum_of_all_elements)
    length_of_list = len(x)
    print("Length of list: ", length_of_list)
    mean_of_list = sum_of_all_elements / length_of_list
    return mean_of_list   

#Function to calculate median of numbers passed in the list
def median_num_friends(x):
    # TODO
    print("List: ", x)
    sorted_list = sorted(x)
    print("Sorted list: ", sorted_list)
    length_of_list = len(x)
    print("Length of list: ", length_of_list)
    mid = (length_of_list-1)//2
    print("Mid: ", mid)
    if (length_of_list % 2 != 0):
        return sorted_list[mid]
    else:
        return ((sorted_list[mid] + sorted_list[mid + 1])/2)       

num_friends = [100, 49, 41, 40, 25, 10, 5, 4, 7, 20, 60]

print("Mean = {}".format(mean_num_friends(num_friends)))
print()
print("Median = {}".format(median_num_friends(num_friends)))
```

```python
from matplotlib import pyplot as plt
import math
import numpy as np

#Function to calculate probabilty density of normal distribution
def normal_pdf(x, mu=0, sigma=1):
    # TODO
    # hits: math.exp
    num = np.exp(-((x-mu)**2))/(2*math.pi*((sigma)**2))
    den = math.sqrt(2*math.pi*((sigma)**2))
    n_pdf = num/den
    return n_pdf

#Showing three graphs when sigma = 1 and sigma = -1
xs = [x / 10.0 for x in range(-50, 50)]
plt.plot(xs, [normal_pdf(x, sigma=1) for x in xs], '-', label='mu=0,sigma=1')
plt.plot(xs, [normal_pdf(x, sigma=2) for x in xs], '-', label='mu=0,sigma=2')
plt.plot(xs, [normal_pdf(x, sigma=0.5) for x in xs], '-', label='mu=0,sigma=0.5')
plt.plot(xs, [normal_pdf(x, sigma=-1) for x in xs], '-', label='mu=0,sigma=-1')
print("Showing three graphs when sigma squared values are equal:")
plt.legend()
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.show()

#Showing four graphs when sigma value is changed
xs = [x / 10.0 for x in range(-50, 50)]
plt.plot(xs, [normal_pdf(x, sigma=1) for x in xs], '-', label='mu=0,sigma=1')
plt.plot(xs, [normal_pdf(x, sigma=2) for x in xs], '-', label='mu=0,sigma=2')
plt.plot(xs, [normal_pdf(x, sigma=0.5) for x in xs], '-', label='mu=0,sigma=0.5')
plt.plot(xs, [normal_pdf(x, sigma=0.3) for x in xs], '-', label='mu=0,sigma=0.3')
print("Showing four graphs when sigma squared values are different:")
plt.legend()
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.show()
```
